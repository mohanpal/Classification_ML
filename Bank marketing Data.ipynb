# Introduction

The dataset is related to marketing campaigns of a Portuguese banking institution. The main objective of the course project is to predict if the target audience/client will enrol for a TERM DEPOSIT or not. This project utilizes “Data Classification” machine learning techniques to uncover hidden characteristics within data. The data consists of 20 input variables with one target variable(y) which defines the client will subscribe “Term deposit” or not(yes/no).

# Data Description

The original data set consists bank marketing information with 41188 instances across 20 inputs which is ordered by date (May’08 to Nov’10). The project is comprised of another data set collected from the same source file. The additional dataset consists of 10% of randomly selected examples of the original collected data i.e. 4119 instances across 20 inputs and one output variable.  

SOURCE of the original DATA – S. Moro, P. Cortez and P. Rita. A Data-Driven Approach to Predict the Success of Bank Telemarketing. Decision Support Systems, Elsevier, 62:22-31, June 2014
SOURCE of the ADDITIONAL/USED data – http://archive.ics.uci.edu/ml/datasets/Bank+Marketing#

### Attributes Information:
1.	Age – The age of the client(numeric)  
2.	Job – Type of the job the client is doing (categorical: 'blue-collar','housemaid','student','management','entrepreneur','retired', 'admin.','self-employed','services','technician','unemployed','unknown')
3.	Marital – What is living marital status of the client (categorical: 'married', 'divorced’, ‘single', 'unknown'; NOTE: Divorced-consists both widowed/divorced)
4.	Education¬ – Till what grade the client has studied (categorical: 'basic.4y','basic.6y','basic.9y', 'illiterate', 'high.school', 'university.degree', 'professional.course', 'unknown')
5.	Default¬ – Is the credit history default or not (categorical: 'yes', 'no', 'unknown')
6.	Housing – Is there a housing loan or not (categorical: 'yes', 'no', 'unknown')
7.	Loan – Is there any loan or not (categorical: 'yes', 'no', 'unknown')
8.	Contact – Type of contact information available (categorical: 'telephone' , 'cellular')
9.	Month – last contact month (categorical: 'jan', ..., 'nov', 'dec')
10.	Day_of_week – Last contact day (categorical: 'mon','tue','wed','thu','fri')
11.	Duration – last contact duration, in seconds (numeric). NOTE: This attribute highly affects the output target (e.g., if duration=0 then y='no'). 

### Social and Economic context attributes

12.	Emp.var.rate – employment variation rate - quarterly indicator (numeric)
13.	Cons.price.idx – consumer price index - monthly indicator (numeric)
14.	Cons.conf.idx – consumer confidence index - monthly indicator (numeric)
15.	Euribor3m –  3 month rate - daily indicator (numeric)
16.	Nr.employed – number of employees - quarterly indicator (numeric)

### Other attributes:
17.	Campaign – number of calls madeto to client (numeric, includes last contact)
18.	Pdays – number of days defining the last contact details of the client (numeric; 999 means client was not previously contacted)
19.	Previous – total number of contacts to specific client (numeric)
20.	Poutcome – output of the previous campaign (categorical: 'failure', 'nonexistent', 'success')



Output variable (desired target) –

21 - y – has the client subscribed to a term deposit (binary: 'yes','no')

#  Predictive Modelling

To get along with machine leaning algorithms, y variable it removed and made to call it as a “Target” while the other features form a “df_part”.

## Partitioning

```python
display(HTML('<b>Table 2: After dropping target variable</b>'))
df_part = df.drop(columns='y')
target = df['y']
target.value_counts()
df_part
```

## ENCODING

Now, encoding the target feature so that the people who subscribe for the term deposit become a positive class as ”1”.

```python
target = target.replace({'no': 0, 'yes': 1})
target.value_counts()
```

As it is clear that the created classes are not balanced. The people accepting for “Term deposit” are only 451 in number as compared to 4119 instances.

```python
df_part['duration'] = df_part['duration'].apply(lambda n:n/60).round(2)
lst = [df_part]
for column in lst:
    column.loc[column["month"] == "jan", "month"] = 1
    column.loc[column["month"] == "feb", "month"] = 2
    column.loc[column["month"] == "mar", "month"] = 3
    column.loc[column["month"] == "apr", "month"] = 4
    column.loc[column["month"] == "may", "month"] = 5
    column.loc[column["month"] == "jun", "month"] = 6
    column.loc[column["month"] == "jul", "month"] = 7
    column.loc[column["month"] == "aug", "month"] = 8
    column.loc[column["month"] == "sep", "month"] = 9
    column.loc[column["month"] == "oct", "month"] = 10
    column.loc[column["month"] == "nov", "month"] = 11
    column.loc[column["month"] == "dec", "month"] = 12
```

## Encoding Categorical features

Since most of the descriptive features are nominal in nature, one-hot-encoding is performed. A dummy variable q is defined for categorical features with q levels. Considering an exception that if a feature has two levels only one variable is defined.


Before transformation , the categorical features are as follows:

```python
categorical_cols = df_part.columns[df_part.dtypes==object].tolist()
categorical_cols
```

For two-level categorical feature:
1. Drop_first function is used to encode variables into single column.
2. Get_dummies function is used to regular one- hot encoding.

```python
display(HTML('<b>Table 3: After One Hot Encoding on dataset</b>'))
# if a categorical descriptive feature has only 2 levels,
# define only one binary variable

for col in categorical_cols:
    n = len(df_part[col].unique())
    if (n == 2):
        df_part[col] = pd.get_dummies(df_part[col], drop_first=True)

# for other categorical features (with > 2 levels), 
# use regular one-hot-encoding 
# if a feature is numeric, it will be untouched

df_part = pd.get_dummies(df_part)
df_part
```

After encoding, following columns are defined:

```python
df_part.columns
```

## Scaling of Features

A min-max scaling of the descriptive features is performed. A copy of data set is made to track the column names:

```python
from sklearn import preprocessing

df_copy = df_part.copy()

Data_scaler = preprocessing.MinMaxScaler()
Data_scaler.fit(df_part)
df_part = Data_scaler.fit_transform(df_part)
```

#  Feature Selection &amp; Ranking

```python
cv_results_full
```

```python
cv_results_full.mean().round(3)
```

```python
feature_importances_rfi = model_rfi.feature_importances_[fs_indices_rfi]
feature_importances_rfi
```

After visualization of these importance’s we get,

```python
%config InlineBackend.figure_format = 'retina'

import altair as alt
alt.renderers.enable('notebook') # required in Jupyter notebooks

def plot_imp(best_features, scores, method_name, color):
    
    data = pd.DataFrame({'features': best_features, 
                       'importances': scores})
    
    chart = alt.Chart(data, 
                      width=500, 
                      title=method_name + ' Feature Importances'
                     ).mark_bar(opacity=0.85, 
                                color=color).encode(
        alt.X('features', title='Feature', sort=None, axis=alt.AxisConfig(labelAngle=45)),
        alt.Y('importances', title='Importance')
    )
    
    return chart
```

# Train-Test splitting

### K-Nearest Neighbors (KNN)

It is the most used algorithms which is non-parametric. The algorithm serves for a data when the points are separated into different classes. On the basis of the classification it predicts a new sample point. The algorithms do not make any assumptions which might be underlying in the data distribution.


#### HOW does it WORK?
It relies upon feature similarity. The algorithm sees how closely input features resembles our training dataset and further determines the classification of the input point.  


#### Limitations:
1. It is computationally expensive as it stores all the training data
2. Sensitive to irrelevant features

#### KNN hyperparameters:
1. (N_neighbors) - Number of neighbors
2. P - Metric distance


```python
from sklearn.base import BaseEstimator, TransformerMixin

# # custom function for RFI feature selection inside a pipeline
# # here we use n_estimators=100
class RFIFeatureSelector(BaseEstimator, TransformerMixin):
    
    # class constructor 
    # make sure class attributes end with a "_"
    # per scikit-learn convention to avoid errors
    def __init__(self, n_features_=10):
        self.n_features_ = n_features_
        self.fs_indices_ = None

    # override the fit function
    def fit(self, X, y):
        from sklearn.ensemble import RandomForestClassifier
        from numpy import argsort
        model_rfi = RandomForestClassifier(n_estimators=100)
        model_rfi.fit(X, y)
        self.fs_indices_ = argsort(model_rfi.feature_importances_)[::-1][0:self.n_features_] 
        return self 
    
    # override the transform function
    def transform(self, X, y=None):
        return X[:, self.fs_indices_]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier

pipe_KNN = Pipeline(steps=[('rfi_fs', RFIFeatureSelector()), 
                           ('knn', KNeighborsClassifier())])

params_pipe_KNN = {'rfi_fs__n_features_': [10, 20, df_part.shape[1]],
                   'knn__n_neighbors': [1, 5, 10, 15, 20],
                   'knn__p': [1, 2]}

gs_pipe_KNN = GridSearchCV(estimator=pipe_KNN, 
                           param_grid=params_pipe_KNN, 
                           cv=cv_method,
                           refit=True,
                           n_jobs=-2,
                           scoring='roc_auc',
                           verbose=1) 
```

```python
def get_search_results(gs):

    def model_result(scores, params):
        scores = {'mean_score': np.mean(scores),
             'std_score': np.std(scores),
             'min_score': np.min(scores),
             'max_score': np.max(scores)}
        return pd.Series({**params,**scores})

    models = []
    scores = []

    for i in range(gs.n_splits_):
        key = f"split{i}_test_score"
        r = gs.cv_results_[key]        
        scores.append(r.reshape(-1,1))

    all_scores = np.hstack(scores)
    for p, s in zip(gs.cv_results_['params'], all_scores):
        models.append((model_result(s, p)))

    pipe_results = pd.concat(models, axis=1).T.sort_values(['mean_score'], ascending=False)

    columns_first = ['mean_score', 'std_score', 'max_score', 'min_score']
    columns = columns_first + [c for c in pipe_results.columns if c not in columns_first]

    return pipe_results[columns]
```

### (Gaussian) Naive Bayes (NB)

It is a powerful algorithm used for predictive modelling. The algorithm relies on the Bayes Theorem 
                                                             P(h|d) = (P(d|h) * P(h)) /  P(d)
##### Where, 

•	P(h|d) is probability of hypothesis (h) in the given data(d)

•	P(d|h) is probability of the data(d) when hypothesis(h) is true

•	P(h) is probability of hypothesis is true

•	P(d) is probability of the data(d)

#### HOW it WORKS?
It predicts the probability of different classes based upon the different attributes. It is an extension of Naive Bayes algorithm where in addition to the probabilities for each class, mean and standard deviation for each input is also stored.

Smoothing is done by using var_smoothing. The value for this parameter is 10^-9. A grid search is done using logspace with a starting value of 10 and  ending value of 10^-3 with 200 values, but a search is performed over only 20 different values.


#### Limitations

•	If the test data set contains a categorical feature which is not observed in training data set, then the model will assign 0 probability (Zero probability). To remove this error we have to use Laplace smoothening.

•	  Is assumes in (independent) pendent predictors which is quite impossible to get in real life.

### Decision Tree (DT)

Decision trees are highly interpretable often used to achieve high accuracy in many tasks. There are four main steps which involves to build a decision tree

•	Training data set should consist of variables and classification 

•	Determination of the best feature in the dataset to split the data

•	(nodes)Splitting the datasets which contains the possible values of the best feature

•	Generating new nodes by using subsets of the data 

#### HOW it WORKS?

In a decision tree internal nodes are taken as test and the leaf nodes can be taken as categories.  These tests are filtered through the tree to get the right output.

#### Limitations

•	Overfitting is the common due to nature of their training 

•	Due to overfitting is becomes biased to some classes which contains majority in the dataset
